```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Car_sales.csv")
df.head()
```

```python
x_axis=df['Manufacturer']
Y_axis=df['Sales_in_thousands']
```

```python
plt.figure(figsize=(20,15))
plt.barh(x_axis, Y_axis)
plt.xlabel('Car Manufacturers')
plt.ylabel('Sales in thousands')
plt.title('Visual analysis of sales')
plt.show()
```
